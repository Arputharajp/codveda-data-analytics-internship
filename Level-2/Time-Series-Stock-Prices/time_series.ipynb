```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

Import the Dataset

```python
df = pd.read_csv("Stock Prices Data Set.csv")
df.head()
```

Inspect Dataset

```python
df.info()
df.shape
```

Convert Date Column to Datetime

```python
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
df.head()
```

Plot Time Series

```python
plt.figure(figsize=(10,5))
plt.plot(df['close'])
plt.title("Stock Price Over Time")
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.show()
```

Moving Average Smoothing

```python
df['MA_30'] = df['close'].rolling(window=30).mean()

plt.figure(figsize=(10,5))
plt.plot(df['close'], label='Original')
plt.plot(df['MA_30'], label='30-Day Moving Average')
plt.title("Stock Price with Moving Average")
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.legend()
plt.show()
```

Time Series Decomposition

```python
decomposition = seasonal_decompose(df['close'], model='additive', period=30)
decomposition.plot()
plt.show()
```

### Observations
- The time series plot shows overall price movement and volatility over time.
- Moving average smoothing highlights the underlying trend by reducing short-term noise.
- Decomposition reveals trend, seasonal patterns, and residual components in the data.

### Conclusion
Time series analysis helped identify trends and seasonality in stock prices, providing insights into temporal patterns and behavior.
