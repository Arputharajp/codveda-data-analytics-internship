Importing Cleaned Dataset

```python
import pandas as pd

df = pd.read_csv(
    "house Prediction Data Set.csv",
    sep=r"\s+",
    header=None
)

df.columns = [
    'CRIM','ZN','INDUS','CHAS','NOX','RM','AGE',
    'DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV'
]

df.info()
df.shape
```

Dataset Overview

```python
df.shape
df.info()
```

Summary Statistics

```python
df.describe()
```

Univariate Analysis

**Histogram**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.histplot(df['MEDV'], kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('House Price ($1000s)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['RM'], kde=True)
plt.title('Distribution of House Area (Average Rooms)')
plt.xlabel('Average Rooms (RM)')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

**Boxplot**

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['MEDV'])
plt.title('Boxplot of House Prices (MEDV)')
plt.xlabel('House Price ($1000s)')
plt.grid(axis='x', alpha=0.75)
plt.show()
```

Bivariate Analysis

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['RM'], y=df['MEDV'])
plt.title('House Area (Average Rooms) vs. House Price')
plt.xlabel('Average Rooms (RM)')
plt.ylabel('House Price ($1000s)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Correlation Analysis

```python
corr = df.corr()
corr
```

Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

### Key Insights
- House prices (MEDV) show a slightly right-skewed distribution with some higher-value outliers.
- The number of rooms (RM) has a strong positive correlation with house prices.
- LSTAT shows a strong negative correlation with MEDV, indicating socioeconomic factors influence pricing.
